# Creating an organic folder and Moving the images

```python
import os
import shutil
import random

source_folder = "/content/drive/MyDrive/Projects/waste_classification_for_recycling/DATASET/TRAIN/O"

destination_folder = "/content/drive/MyDrive/Projects/waste_classification_for_recycling/Data/Garbage classification/Garbage classification/organic"

# List of image file extensions to filter by
image_extensions = ['.jpg', '.jpeg', '.png', '.gif']

# Get a list of image files in the source folder
image_files = [filename for filename in os.listdir(source_folder) if any(filename.lower().endswith(ext) for ext in image_extensions)]

# Shuffle the list of image files
random.shuffle(image_files)

# Counter to keep track of the number of files moved
files_moved = 0

# Loop through the shuffled list of image files and move them to the destination folder
for filename in image_files:
    if files_moved >= 1000:
        break  # Stop after moving 1000 files
    source_file_path = os.path.join(source_folder, filename)
    destination_file_path = os.path.join(destination_folder, filename)
    shutil.move(source_file_path, destination_file_path)
    files_moved += 1

print(f"{files_moved} shuffled images moved successfully.")
```
